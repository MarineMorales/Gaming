# Game: Dragon Realm

## 1. Rules

The player is in a land full of dragons.
The dragons all live in caves with their large piles of collected treasure.
Some dragons are friendly and share their treasure.
Other dragons are hungry and eat anyone who enters their cave.
The player approaches two caves, one with a friendly dragon and the other with a hungry dragon, but doesn't know which dragon is in the cave.
The players decides between the two.

## 2. Flowchart

```python
from graphviz import Digraph
```

```python
dot = Digraph(comment='Dragon Realm Flowchart')

dot.node('START', 'START', shape='oval', style='filled', fillcolor='grey')
dot.node('Display Intro', 'Display Intro')
dot.node('Choose Cave', 'Choose Cave')
dot.node('Check Cave', 'Check Cave')

friendly_dragon = Digraph('Friendly Dragon', node_attr={'shape': 'doublecircle', 'color': 'green'})
friendly_dragon.node('Win','Win')
dot.subgraph(friendly_dragon)

hungry_dragon = Digraph('Hungry Dragon', node_attr={'shape': 'doublecircle', 'color': 'red'})
hungry_dragon.node('Lose','Lose')
dot.subgraph(hungry_dragon)

dot.node('Ask to play again', 'Ask to play again')
dot.node('END', 'END', shape='oval', style='filled', fillcolor='grey')
```

```python
dot.edge('START', 'Display Intro')
dot.edge('Display Intro', 'Choose Cave')
dot.edge('Choose Cave', 'Check Cave')
dot.edge('Check Cave', 'Win', label='Friendly Dragon')
dot.edge('Check Cave', 'Lose', label='Hungry Dragon')
dot.edge('Win', 'Ask to play again')
dot.edge('Lose', 'Ask to play again')
dot.edge('Ask to play again', 'Display Intro', label='Yes')
dot.edge('Ask to play again', 'END', label='No')
```

## 3. Program

```python
import random
import time
```

```python
def displayIntro():
    print('''You are in a land full of dragons.
    In front of you, you see two caves.
    In one cave, the dragon is friendly and will share his treasure with you.
    The other dragon is greedy and hungry, and will eat you on sight.''')
    # Multiple strings: if you use 3 quotes at the start and the end of a string, then it can go accross several lines.
    print()
    
def chooseCave():
    cave = ''
    while cave!='1' and cave!='2': 
        # The while loop will go on until you enter either 1 or 2 
        print('Which cave will you go into? (1 or 2)')
        cave = input()
    return cave
    # Now your variable is not only defined in the chooseCave local scope but also in the program global scope.

def checkCave(chosenCave):
    # ChosenCave is a local variable parameter
    print('You approach the cave ...')
    time.sleep(2)
    print('It is dark and spooky...')
    time.sleep(2)
    print('A large dragon jumps out in front of you! He opens his jaws and ...')
    print()
    time.sleep(2)
    
    friendlyCave = random.randint(1,2)
    
    if chosenCave == str(friendlyCave):
        print('Gives you his treasure!')
    else:
        print('Gobbles you down in one bite!')
```

```python
playAgain = 'yes'
while playAgain == 'yes':
    displayIntro()
    caveNumber = chooseCave()
    checkCave(caveNumber)
    
    print('Do you want to play again? (yes or no)')
    playAgain = input()
```
